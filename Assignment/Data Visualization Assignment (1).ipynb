```python
#values of the sine wave
x = np.arange(0, 10, 0.1);
amp = np.sin(x)
```

```python
plt.plot(x,amp,'k-.')
plt.axhline(y=0, color='k')
plt.xlabel('X',size=10)
plt.ylabel('Y',size=10)
plt.title('Sine Wave',size=20)
```
